<h2>Libraries import</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Dataset import and preparation</h2>

```python
#import and display data

pd.options.display.max_columns = None
pd.options.display.max_rows = None
df = pd.read_csv('./csgo_round_snapshots.csv')

df = df.sample(1000, random_state = 42)
df.shape
```

```python
#factorize categorical values to numerical (str ~> int)
df['map'] = pd.factorize(df['map'])[0]
df['bomb_planted'] = pd.factorize(df['bomb_planted'])[0]
df['round_winner'] = pd.factorize(df['round_winner'])[0]
```

```python
#drop unnecessary features
X = df.drop(['round_winner','t_money', 'ct_money', 'ct_grenade_decoygrenade', 't_grenade_decoygrenade'],axis=1) #t_money and ct_money dropped performance by 1%

#combine smgs in one column + sum for team
t_smg_names = ['t_weapon_bizon', 't_weapon_mac10', 't_weapon_mp5sd', 't_weapon_mp7', 't_weapon_mp9', 't_weapon_p90', 't_weapon_ump45']
ct_smg_names = ['ct_weapon_bizon', 'ct_weapon_mac10', 'ct_weapon_mp5sd', 'ct_weapon_mp7', 'ct_weapon_mp9', 'ct_weapon_p90', 'ct_weapon_ump45']
X['t_weapon_smgs']= df[t_smg_names].sum(axis=1)
X['ct_weapon_smgs']= df[ct_smg_names].sum(axis=1)

#combine shotguns in one column + sum for team
t_shotgun_names = ['t_weapon_nova', 't_weapon_mag7', 't_weapon_sawedoff', 't_weapon_xm1014']
ct_shotgun_names = ['ct_weapon_nova', 'ct_weapon_mag7', 'ct_weapon_sawedoff', 'ct_weapon_xm1014']
X['t_weapon_shotguns']= df[t_shotgun_names].sum(axis=1)
X['ct_weapon_shotguns']= df[ct_shotgun_names].sum(axis=1)

#combine lmgs in one column + sum for team
t_lmg_names = ['t_weapon_m249', 't_weapon_negev']
ct_lmg_names = ['ct_weapon_m249', 'ct_weapon_negev']
X['t_weapon_lmgs']= df[t_lmg_names].sum(axis=1)
X['ct_weapon_lmgs']= df[ct_lmg_names].sum(axis=1)

#drop
X = X.drop(t_smg_names + ct_smg_names + t_shotgun_names + ct_shotgun_names + t_lmg_names + ct_lmg_names, axis=1)
y = df['round_winner'].values
print(f"Number of features : {len(X.columns)}")
X.head()
```

<h2>Model Training</h2>
